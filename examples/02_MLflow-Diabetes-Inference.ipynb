```python
import mlflow
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
```

Open your experiment and your chosen run in your MLflow Tracking server. Under artifacts is model directory, where Full Path of model uri can be copied. Paste it into the code in the next cell.



```python
# Load and prepare data and model 
db = load_diabetes()
test_data = db.data

model_path = "your_model_path"
model = mlflow.sklearn.load_model(model_path)

# Predict on loaded model
predictions = model.predict(test_data)
#print(predictions)
```

```python
# Create dataframe to compare predicted and true values of glucose levels

data = {
    'Predicted Glucose Levels': predictions,
    'Actual Glucose Levels': db.target
    }

df = pd.DataFrame(data)
print(df)
```
